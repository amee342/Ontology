### Paper 1 SHARPIE

```python
with onto:

  # ================= SCENARIO =================
  scenario_sharpie = onto.Scenario("SHARPIE_HumanAI_RL_Scenario")

  domain_sharpie = onto.Domain("ReinforcementLearning")

  context_sharpie_1 = onto.Context("HybridHumanAIInteraction")
  context_sharpie_2 = onto.Context("SequentialDecisionMaking")
  context_sharpie_3 = onto.Context("MultimodalCommunication")

  endgoal_sharpie = onto.Endgoal("StandardizeHumanAIExperiments")

  ethical_sharpie_1 = onto.EthicalConsideration("ResponsibleAIUse")
  ethical_sharpie_2 = onto.EthicalConsideration("HumanInvolvement")

  # property assertions
  scenario_sharpie.domain = [domain_sharpie]
  scenario_sharpie.context = [
      context_sharpie_1,
      context_sharpie_2,
      context_sharpie_3,
  ]
  scenario_sharpie.endgoal = [endgoal_sharpie]
  scenario_sharpie.hasEthicalConsideration = [
      ethical_sharpie_1,
      ethical_sharpie_2,
  ]

  # ================= ACTORS =================

  human_participant = onto.Human("HumanParticipant")
  rl_agent = onto.ArtificialAgent("RL_Agent")

  human_participant.inScenario = [scenario_sharpie]
  rl_agent.inScenario = [scenario_sharpie]

  # capabilities of RL agent (based on paper)
  rl_agent.capability = [
      onto.Learning,
      onto.Adaptiveness,
      onto.DecisionMaking,
  ]

  # ================= INTERACTION =================

  interaction_sharpie = onto.Interaction("Interaction_Human_RLAgent")
  interaction_sharpie.interactingAgent = [
      human_participant,
      rl_agent,
  ]

  interaction_sharpie.interactionTask = [
      onto.Collaboration
  ]

  interaction_sharpie.interactionMethod = [
      onto.Multimodal
  ]

  scenario_sharpie.hasInteraction = [interaction_sharpie]

  # ================= INFORMATION PROCESSING =================

  rl_processing = onto.InformationProcessing("RLAgent_Processing")

  rl_processing.processingTask = [
      onto.Learning,          # agent learns from environment
      onto.Reasoning,         # sequential decision making
      onto.Transforming,      # actions in environment
  ]

  rl_processing.informationMethod = [
      onto.Statistical
  ]

  rl_agent.processingInformation = [rl_processing]
```

### Paper 2 HILO

```python
with onto:

  # ================= SCENARIO =================
  scenario_hilo = onto.Scenario("HILO_LegalDesk_Scenario")

  domain_hilo = onto.Domain("LearningOrganization")

  context_hilo_1 = onto.Context("KnowledgeErosion")
  context_hilo_2 = onto.Context("HumanAICoEvolution")
  context_hilo_3 = onto.Context("DecisionSupportConsultation")

  endgoal_hilo = onto.Endgoal("EnhanceOrganizationalIntelligence")

  ethical_hilo_1 = onto.EthicalConsideration("HumanOversight")
  ethical_hilo_2 = onto.EthicalConsideration("Explainability")

  # property assertions
  scenario_hilo.domain = [domain_hilo]
  scenario_hilo.context = [
      context_hilo_1,
      context_hilo_2,
      context_hilo_3,
  ]
  scenario_hilo.endgoal = [endgoal_hilo]
  scenario_hilo.hasEthicalConsideration = [
      ethical_hilo_1,
      ethical_hilo_2,
  ]

  # ================= ACTORS =================

  # workers (novice and expert legal advisors)
  legal_advisor_novice = onto.Worker("LegalAdvisor_Novice")
  legal_advisor_expert = onto.Worker("LegalAdvisor_Expert")

  legal_ai = onto.ArtificialAgent("LegalDecisionSupportAI")

  legal_advisor_novice.inScenario = [scenario_hilo]
  legal_advisor_expert.inScenario = [scenario_hilo]
  legal_ai.inScenario = [scenario_hilo]

  # capabilities of AI (based on paper)
  legal_ai.capability = [
      onto.Reasoning,
      onto.Learning,
      onto.DecisionMaking,
  ]

  # ================= INTERACTION =================

  interaction_hilo = onto.Interaction("Interaction_LegalAdvisors_AI")
  interaction_hilo.interactingAgent = [
      legal_advisor_novice,
      legal_advisor_expert,
      legal_ai,
  ]

  interaction_hilo.interactionTask = [
      onto.DecisionSupport
  ]

  interaction_hilo.interactionMethod = [
      onto.HumanAIfeedbackLoop
  ]

  scenario_hilo.hasInteraction = [interaction_hilo]

  # ================= INFORMATION PROCESSING =================

  ai_processing_hilo = onto.InformationProcessing("LegalAI_Processing")

  ai_processing_hilo.processingTask = [
      onto.Reasoning,     # analyze cases
      onto.Learning,      # improve from feedback
      onto.Transforming,  # generate advice
  ]

  ai_processing_hilo.informationMethod = [
      onto.Statistical
  ]

  legal_ai.processingInformation = [ai_processing_hilo]
```

### Paper 3 BRIDGET

```python
with onto:

  # ================= SCENARIO =================
  scenario_bridget = onto.Scenario("BRIDGET_HybridDecisionScenario")

  domain_bridget = onto.Domain("HybridDecisionMaking")

  context_bridget_1 = onto.Context("HumanInTheLoopLabeling")
  context_bridget_2 = onto.Context("LearningToDefer")
  context_bridget_3 = onto.Context("SkepticalLearning")

  endgoal_bridget = onto.Endgoal("BridgeHybridDecisionGap")

  ethical_bridget_1 = onto.EthicalConsideration("Explainability")
  ethical_bridget_2 = onto.EthicalConsideration("HumanControl")

  # property assertions
  scenario_bridget.domain = [domain_bridget]
  scenario_bridget.context = [
      context_bridget_1,
      context_bridget_2,
      context_bridget_3,
  ]
  scenario_bridget.endgoal = [endgoal_bridget]
  scenario_bridget.hasEthicalConsideration = [
      ethical_bridget_1,
      ethical_bridget_2,
  ]

  # ================= ACTORS =================

  human_labeler = onto.Human("HumanLabeler")
  bridget_ai = onto.ArtificialAgent("BRIDGET_System")

  human_labeler.inScenario = [scenario_bridget]
  bridget_ai.inScenario = [scenario_bridget]

  # capabilities of BRIDGET (supported by paper)
  bridget_ai.capability = [
      onto.Learning,
      onto.Reasoning,
      onto.DecisionMaking,
      onto.Adaptiveness,
  ]

  # ================= INTERACTION =================

  interaction_bridget = onto.Interaction("Interaction_Human_BRIDGET")
  interaction_bridget.interactingAgent = [
      human_labeler,
      bridget_ai,
  ]

  interaction_bridget.interactionTask = [
      onto.DataAnnotation
  ]

  interaction_bridget.interactionMethod = [
      onto.AdaptiveControlSwitching
  ]

  scenario_bridget.hasInteraction = [interaction_bridget]

  # ================= INFORMATION PROCESSING =================

  bridget_processing = onto.InformationProcessing("BRIDGET_Processing")

  bridget_processing.processingTask = [
      onto.Reasoning,     # prediction & uncertainty
      onto.Learning,      # incremental updates
      onto.Transforming,  # produce labels/decisions
  ]

  bridget_processing.informationMethod = [
      onto.Statistical
  ]

  bridget_ai.processingInformation = [bridget_processing]
```
